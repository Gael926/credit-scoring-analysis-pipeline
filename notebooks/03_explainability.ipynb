# 03. Explicabilité (SHAP)

Ce notebook génère les graphiques d'explicabilité pour le meilleur modèle entraîné.

```python
import pandas as pd
import joblib
import sys
import os
import matplotlib.pyplot as plt

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path: sys.path.append(project_root)

from src.explainability import compute_shap, plot_shap_global, plot_shap_local

plt.style.use('ggplot')
```

```python
# Charger modèle
model_path = "../models/best_model.pkl"
if not os.path.exists(model_path):
    raise FileNotFoundError("Le modèle n'existe pas. Lancez le notebook 02 d'abord.")
model = joblib.load(model_path)

# Charger data (on charge V2 par défaut, ou on pourrait automatiser)
# Idéalement on garde une trace de quel dataset a gagné, mais ici on charge V2 pour l'exemple
X = pd.read_pickle('../data/processed/X_prepared_v1.pkl')
X.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in X.columns]
X_sample = X.sample(1000, random_state=42)
```

```python
save_dir = "../reports/figures"
explainer, shap_values = compute_shap(model, X_sample)

# Plot Global
plot_shap_global(shap_values, X_sample, save_path=f"{save_dir}/shap_global.png")
```

```python
# Plot Local pour quelques clients
for i in [0, 5, 10]:
    plot_shap_local(explainer, X_sample, i, save_path=f"{save_dir}/shap_local_{i}.png")
```
